# Membuat Visualisasi dengan Seaborn

## Seaborn?

> Seaborn merupakan Library di Python yang khusus untuk membuat plot statistik. Dibuat berdasarkan library Matplotlib 
dan sangat cocok dipakai bersama library Pandas

### Import Module

```python
import pandas as pd
import seaborn as sns
sns.set()
```

### Persiapan Data

```python
dataku = pd.read_csv('titanic.csv')
dataku.head()
```

### Membuat Distribution Plot

```python
sns.distplot(dataku.age); # .distplot() Sudah tidak lagi di maintenance sehingga diganti dengan .displot() # Warning Hilang
```

### Menghilangkan Warning

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
sns.displot(dataku.age);
```

```python
sns.displot(dataku.age, bins=50); # bins menampilkan jumlah hist yang ditampilkan
```

```python
sns.jointplot(x='age', y='fare', data=dataku);
```

```python
sns.jointplot('age', 'fare', kind='reg', data=dataku);
```

### Membuang Outlier

#### Cara Manual

```python
dataku.fare[dataku.fare>400]
```

```python
dataku_new = dataku.drop([49,50,183,302])
```

#### Cek Kembali

```python
sns.jointplot('age', 'fare', kind='reg', data=dataku_new);
```

```python
sns.jointplot('age', 'fare', kind='hex', data=dataku);
```

```python
sns.jointplot('age', 'fare', kind='kde', data=dataku);
```

### Membuat Pair Plot

```python
sns.pairplot(dataku);
```

#### Menambah Argumen pada Pair Plot

```python
sns.pairplot(dataku, hue='age');
```

```python
sns.pairplot(dataku, hue='sex');
```

## Membuat Seaborn Kategori

### Import Module

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

### Persiapan Data

```python
dataku = pd.read_csv('titanic.csv')
```

```python
dataku.head()
```

```python
dataku.describe(include='O')
```

### Visualisasi dengan Bar Plot (Histogram)

```python
sns.barplot(x='sex', y='fare', data=dataku);
```

#### Merubah Orientasi

```python
sns.barplot(x='fare', y='sex', data=dataku);
```

#### Merubah Estimater

```python
sns.barplot(x='sex', y='fare', data=dataku, estimator=sum);
```

```python
sns.barplot(x='sex', y='fare', data=dataku, estimator=np.std);
```

### Visualisasi dengan Count Plot

```python
sns.countplot(x='sex', data=dataku);
```

```python
sns.countplot(x='embarked', hue='sex', data=dataku);
```

#### Mengurutkan Data

```python
sns.countplot(x='embarked', order=['C','S','Q'], hue='sex', data=dataku);
```

### Visualisasi dengan Box Plot

```python
sns.boxplot(x='sex', y='fare', data=dataku);
```

```python
sns.boxplot(x='sex', y='fare', hue='embarked', data=dataku);
```

### Visualisasi dengan Violin Plot

```python
sns.violinplot(x='sex', y='fare', data=dataku);
```

```python
sns.violinplot(x='sex', y='fare', hue='embarked', data=dataku);
```

```python
sns.violinplot(x='embarked', y='fare', hue='sex', split=True, data=dataku);
```

### Visualisasi dengan Strip Plot

```python
sns.stripplot(x='embarked', y='fare', hue='sex', jitter=True, data=dataku);
```

## Membuat Matrix Plot

### Import Module

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
dataku = pd.read_csv('titanic.csv')
```

```python
dataku.head()
```

### Membuat Matrix Heatmap

```python
dataku.corr()
```

```python
sns.heatmap(dataku.corr());
```

```python
sns.heatmap(dataku.corr(), annot=True);
```

```python
sns.heatmap(dataku.corr(), annot=True, cmap='Reds')
plt.yticks(rotation=0);
```

```python
sns.heatmap(dataku.corr(), annot=True, cmap='Greens')
plt.yticks(rotation=0);
```

### Membuat Heatmap dengan Tabel Data Frame Apa Adanya

```python
dataku2 = pd.read_csv('titanic.csv', index_col='embarked')
```

```python
dataku2.head()
```

```python
sns.heatmap(dataku2.iloc[:,[0,1,5,6]].head(), annot=True, cmap='Greens');
```
