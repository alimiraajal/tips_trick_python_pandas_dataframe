# Konsep Dasar Matplotlib

## Membuat Grafik dengan metode Konvensional

### Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Kanvas 

```python
plt.plot()
```

### Nilai .plot()

```python
plt.plot([10,20,30,40,50])
plt.show()
```

```python
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Judul Grafik')
plt.show()
```

### Membuat 2 Grafik Berdampingan dengan .subplot()

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='red')
plt.xlabel('Sumbu X1')
plt.ylabel('Sumbu Y1')

plt.subplot(1,2,2)
plt.plot(y,x, color='blue')
plt.xlabel('Sumbu X2')
plt.ylabel('Sumbu Y2')

plt.tight_layout()
plt.show()
```

## Membuat Grafik dengan Metode Object Oriented 

### Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Persiapan Data

```python
x = np.arange(1, 20, 0.5)
y = x**2
```

```python
x
```

```python
y
```

### Versi 1 OOP Matplotlib

```python
fig = plt.figure() # menyiapkan kanvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # menampilkan kanvas axes sebagai OOP
axes.plot(x,y) # menampilkan grafik berdasarkan x dan y
axes.set_xlabel('Sumbu X') # axes.set_ digunakan untuk melakukan pengaturan grafik
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik')
plt.show()
```

### Menambahkan Axes Lain di dalam Axes

```python
fig = plt.figure() # menyiapkan kanvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # menampilkan kanvas axes1 sebagai OOP
axes2 = fig.add_axes([0.2,0.5,0.3,0.3]) # menampilkan kanvas axes1 sebagai OOP

axes1.plot(x,y)
axes1.set_xlabel('Sumbu X')
axes1.set_ylabel('Sumbu Y')
axes1.set_title('Grafik Utama')

axes2.plot(y,x)
axes2.set_xlabel('Sumbu X')
axes2.set_ylabel('Sumbu Y')
axes2.set_title('Grafik Tambahan')
plt.show()
```

### Versi 2 OOP Matplotlib

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_xlabel('Sumbu X')
axes[0].set_ylabel('Sumbu Y')
axes[0].set_title('Judul Grafik 1')

axes[1].plot(y,x)
axes[1].set_xlabel('Sumbu X')
axes[1].set_ylabel('Sumbu Y')
axes[1].set_title('Judul Grafik 2')
fig.tight_layout()
plt.show()
```

## Pengaturan Grafik Matplotlib

### Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Persiapan Data

```python
x = np.arange(1,20,0.5)
y = x**2
```

### Atur DPI - OOP versi 1

```python
fig = plt.figure(figsize=(4,4), dpi=200) # .figsize() untuk mengatur ukuran saat dicetak (satuan Inches) DPI maksimal 300
axes = fig.add_axes([0,0,1,1]) # buat OOP dengan nama axes
axes.plot(x,y)
plt.show()
```

### Atur DPI - OOP Versi 2

```python
fig, axes = plt.subplots(2,1, figsize=(4,4), dpi=200) # ukuran figsize merupakan ukuran canvas secara keseluruhan

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
plt.show()
```

### Menyimpan Gambar Grafik

```python
fig.savefig('gambar_grafik.png', dpi=300) # Isi nama file gambar beserta format (rekomendasi PNG) dan dapat mengatur DPI
```

### Menambahkan Legend (Keterangan grafik)

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y**2, label='Y pangkat 2')
axes.plot(x,x**2, label='X pangkat 2')
axes.plot(x,x**3, label='X pangkat 3')
axes.legend(loc='best')
plt.show()
```

### Menambah Marker, Line Width, dan Line Style

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.1,0.4,0.3,0.3])

axes.plot(x,y**2, label='Y pangkat 2', linewidth=2, linestyle=':', marker='d')
axes.plot(x,x**2, label='X pangkat 2', linewidth=2, linestyle='-', marker='+')
axes.plot(x,x**3, label='X pangkat 3', linewidth=2, linestyle='-.', marker='o')

axes1.plot(x,y, linewidth=1, linestyle='-')
axes1.set_xlabel('Sumbu X')
axes1.set_ylabel('Sumbu Y')
axes1.set_title('Grafik Tambahan')

axes.legend(loc='best')
plt.show()
```
